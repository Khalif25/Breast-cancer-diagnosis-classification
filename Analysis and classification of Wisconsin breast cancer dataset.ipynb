```python
# check distribution of values in the column
# Determine the number of rows and columns for subplots
num_cols = len(df.columns)
num_rows = (num_cols - 1) // 3 + 1  # Ensure at least 1 row

# Plotting
fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(20, 6 * num_rows))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot each column
for i, (column, ax) in enumerate(zip(df.columns, axes)):
    ax.hist(df[column].dropna(), bins=10, edgecolor='black')
    ax.set_title(column)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.grid(True)

# Hide empty subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
